```python
# Lấy danh sách các nhà máy theo thứ tự xuất hiện đầu tiên
ordered_plants = df_TDN['TEN_NM'].drop_duplicates().tolist()

# Dictionary để lưu trữ DataFrame cho từng nhà máy
plant_dfs = {}

count = 0
# Tạo DataFrame riêng cho từng nhà máy
for plant in ordered_plants:
    # Lọc dữ liệu theo tên nhà máy
    plant_df = df_TDN[df_TDN['TEN_NM'] == plant].copy()
    
    # Lưu DataFrame vào dictionary với key là tên nhà máy
    plant_dfs[plant] = plant_df
    
    # Hiển thị thông tin về DataFrame của mỗi nhà máy (tùy chọn)
    print(f"Nhà máy: {plant}, Số dòng: {len(plant_df)}")
    
    count += 1

# Bây giờ bạn có thể truy cập từng DataFrame riêng biệt thông qua plant_dfs
# Ví dụ: plant_dfs['Tên nhà máy cụ thể']
```

```python
# Lấy DataFrame của nhà máy NAM_CUN
nam_cun_df = plant_dfs["NAM_CUN"].copy()

# Kiểm tra các MADIEMDO khác nhau trong NAM_CUN
unique_madiemdo = nam_cun_df['MADIEMDO'].unique()
print(f"Các MADIEMDO khác nhau trong NAM_CUN: {unique_madiemdo}")
print(f"Số lượng MADIEMDO: {len(unique_madiemdo)}")

# Lấy danh sách các cột thời gian từ "00:00:00" đến "23:50:00"
time_columns = [col for col in nam_cun_df.columns if ':' in col]
print(f"Tổng số cột thời gian: {len(time_columns)}")

# Nhóm theo NGAY và tính tổng các cột thời gian
# Điều này sẽ gộp tất cả các MADIEMDO khác nhau trong cùng một ngày
result_df = nam_cun_df.groupby(['NGAY'])[time_columns].sum().reset_index()

# Thêm thông tin nhà máy
result_df['TEN_NM'] = "NAM_CUN"

# Thêm cột tổng ngày (tổng tất cả các giá trị thời gian trong ngày)
result_df['TONG_NGAY'] = result_df[time_columns].sum(axis=1)

# Hiển thị thông tin kết quả
print(f"\nDữ liệu trước khi gộp: {nam_cun_df.shape[0]} dòng")
print(f"Dữ liệu sau khi gộp: {result_df.shape[0]} dòng")

# Hiển thị một phần của DataFrame kết quả
print("\nNăm dòng đầu tiên của kết quả:")
print(result_df[['NGAY', 'TEN_NM', 'TONG_NGAY', '00:00:00', '00:30:00', '01:00:00']].head())

# Lưu kết quả để kiểm tra
nam_cun_aggregated = result_df

# Kiểm tra tổng của một ngày cụ thể trước và sau khi gộp
if not nam_cun_df.empty and not result_df.empty:
    # Chọn một ngày làm mẫu
    sample_day = nam_cun_df['NGAY'].iloc[3]
    
    # Tính tổng cho ngày đó trước khi gộp
    before_agg = nam_cun_df[nam_cun_df['NGAY'] == sample_day][time_columns].sum().sum()
    
    # Tìm tổng cho ngày đó sau khi gộp
    after_agg = result_df[result_df['NGAY'] == sample_day]['TONG_NGAY'].iloc[0]
    
    print(f"\nKiểm tra tổng cho ngày {sample_day}:")
    print(f"Tổng trước khi gộp: {before_agg}")
    print(f"Tổng sau khi gộp: {after_agg}")
    
    if abs(before_agg - after_agg) < 0.01:
        print("Kết quả gộp chính xác!")
    else:
        print("Có sự khác biệt trong kết quả!")
```
