```python
import pandas as pd
import unidecode
import re

def convert_nmtd_to_ten_nm(text):
    """
    - Bỏ tiền tố 'NMTĐ' hoặc 'NMTD' ở đầu (nếu có).
    - Loại bỏ dấu tiếng Việt.
    - Thay khoảng trắng và các ký tự đặc biệt thành dấu gạch dưới '_'.
    - Xóa bớt gạch dưới thừa ở đầu/cuối.
    """
    if pd.isna(text):
        return ""

    # Bỏ tiền tố 'NMTĐ' hoặc 'NMTD' (không phân biệt hoa thường)
    text = re.sub(r'^(?:NMTĐ|NMTD)\s*', '', text, flags=re.IGNORECASE)

    # Loại bỏ dấu tiếng Việt
    text = unidecode.unidecode(text)

    # Thay các ký tự không phải chữ hoặc số (kể cả dấu cách, dấu gạch ngang, dấu chấm...) thành '_'
    text = re.sub(r'[^A-Za-z0-9]+', '_', text)

    # Xóa dấu gạch dưới ở đầu hoặc cuối
    text = text.strip('_')

    # Chuyển về in hoa hoặc in thường tùy ý, ở đây ví dụ để in hoa:
    text = text.upper()

    return text

# Đọc dữ liệu từ file Excel
df = pd.read_excel(r"C:\Khue\TDN\data\raw\thong_tin_NMTD_main.xlsx")

# Tạo cột mới 'TEN_NM' từ cột 'NMTD'
df['TEN_NM'] = df['NMTD'].apply(convert_nmtd_to_ten_nm)

df.to_excel(r"C:\Khue\TDN\data\raw\thong_tin_NMTD_convert_ver2.xlsx", index=None)
```

```python
import pandas as pd
import unidecode
import re

def convert_nmtd_to_ten_nm(text):
    """
    1. Chuyển về in hoa.
    2. Loại bỏ dấu tiếng Việt (dùng unidecode).
    3. Xóa tiền tố "NMTD" hoặc "MNTD" (sau khi bỏ dấu).
    4. Thay các ký tự không phải chữ/số thành dấu gạch dưới '_'.
    5. Xóa dấu gạch dưới ở đầu/cuối.
    """
    if pd.isna(text):
        return ""

    # 1. Chuyển về in hoa
    text = text.upper()

    # 2. Loại bỏ dấu tiếng Việt (chuyển NMTĐ, MNTĐ thành NMTD, MNTD...)
    text = unidecode.unidecode(text)

    # 3. Xóa tiền tố "NMTD" hoặc "MNTD" nếu có
    #    (?:...) là nhóm không capture, giúp gọn regex
    text = re.sub(r'^(?:NMTD|MNTD)\s*', '', text)

    # 4. Thay các ký tự không phải chữ/số bằng '_'
    text = re.sub(r'[^A-Za-z0-9]+', '_', text)

    # 5. Xóa dấu gạch dưới ở đầu hoặc cuối
    text = text.strip('_')

    return text

# Ví dụ kiểm tra nhanh
test_strings = [
    "MNTĐ KHE THƠI",
    "NMTÐ NẬM NA 1",
    "NMTĐ BẢO LÂM 1",
    "MNTD Nậm Mô",
]
for s in test_strings:
    print(s, "->", convert_nmtd_to_ten_nm(s))

# Tích hợp vào file Excel
df = pd.read_excel(r"C:\Khue\TDN\data\raw\thong_tin_NMTD_main.xlsx")
df["TEN_NM"] = df["NMTD"].apply(convert_nmtd_to_ten_nm)
df.to_excel(r"C:\Khue\TDN\data\raw\thong_tin_NMTD_convert_ver2.xlsx", index=False)
print("Đã chuyển đổi và lưu kết quả.")
```
