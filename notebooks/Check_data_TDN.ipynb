```python
import re

df_list = []   

for file in os.listdir(folder_path):
    if file.endswith('.xlsx'):
        # Extract date from filename using regular expression
        date_match = re.search(r'(\d{4}\.\d{2}\.\d{2})', file)
        date_part = date_match.group(1) if date_match else ""
        
        if '-A1' in file:
            # Đọc dữ liệu từ 'Báo cáo NMĐ QĐK A1' 
            df_A1 = pd.read_excel(os.path.join(folder_path, file), skiprows=162, usecols='A:BC', nrows=860, header=0)
            # Thêm cột 'Ngày' vào sau cột 'Loại'
            df_A1.insert(7, 'Ngày', date_part)
            df_list.append(df_A1)  # Thêm df_A1 vào danh sách
        
        # Kiểm tra nếu file có tên chứa '-NPC'
        elif '-NPC' in file:
            # Đọc dữ liệu từ 'Báo cáo NMĐ QĐK PC' 
            df_NPC = pd.read_excel(os.path.join(folder_path, file), skiprows=92, usecols='A:BC', nrows=726, header=0)
            # Thêm cột 'Ngày' vào sau cột 'Loại'
            df_NPC.insert(7, 'Ngày', date_part)
            df_list.append(df_NPC)  # Thêm df_NPC vào danh sách
            
# Kết hợp tất cả các DataFrame lại
final_df = pd.concat(df_list, ignore_index=True)
```

```python
import unicodedata

def normalize_vietnamese(text):
    """
    Chuẩn hóa chuỗi Unicode về dạng NFC.
    NFC (Normalization Form C) 'gộp' các dấu phụ về ký tự cơ sở,
    giúp hạn chế lỗi do tồn tại ký tự ghép hoặc ký tự combining.
    
    Handles non-string inputs by converting them to strings first.
    """
    if not isinstance(text, str):
        text = str(text)
    return unicodedata.normalize('NFC', text)

# Chuẩn hóa dữ liệu trong cột 'Nhà máy' của final_df
final_df['Nhà máy'] = final_df['Nhà máy'].apply(normalize_vietnamese)
```

```python
def format_plant_name(name):
    # Loại bỏ "NMTĐ " hoặc "NMTD " ở đầu (nếu có)
    if 'NMTĐ ' in name:
        name = name.replace('NMTĐ ', '')
    elif 'NMTD ' in name:
        name = name.replace('NMTD ', '')
    
    # Chuyển tất cả về lowercase
    name = name.lower()
    
    # Danh sách các từ viết tắt cần giữ nguyên dạng viết hoa
    abbr = ['3a', '3b', '1a', '8b']
    
    # Tách thành list các từ
    words_list = name.split()
    
    # Xử lý từng từ
    words = []
    for i, word in enumerate(words_list):
        # Kiểm tra nếu là từ cuối cùng và là "a" hoặc "b"
        if i == len(words_list) - 1 and word.lower() in ['a', 'b']:
            words.append(word.upper())
        # Kiểm tra nếu từ có trong danh sách viết tắt
        elif word in abbr:
            words.append(word.upper())
        else:
            words.append(word.capitalize())
    
    return ' '.join(words)

# Áp dụng hàm chuyển đổi
df_check['NMTD'] = df_check['NMTD'].apply(format_plant_name)

# Hiển thị kết quả
print(df_check['NMTD'].head(10))
```

```python
def loc_du_lieu_2024(df, thang, ten_don_vi_dien_luc):
    # Lọc dữ liệu theo tên đơn vị điện lực
    df_filtered = df[df['DVDL'] == ten_don_vi_dien_luc].copy()
    
    # Chuyển cột 'NGAY' về dạng datetime nếu chưa đúng định dạng
    df_filtered['NGAY'] = pd.to_datetime(df_filtered['NGAY'], format='%Y-%m-%d', errors='coerce')
    
    # Lọc theo tháng 8 và năm 2024
    df_filtered = df_filtered[
        (df_filtered['NGAY'].dt.month == thang) &
        (df_filtered['NGAY'].dt.year == 2024)
    ].copy()

    # Reset lại index
    df_filtered.reset_index(drop=True, inplace=True)

    return df_filtered
```

```python
import re
import numpy as np
from difflib import get_close_matches

# Hàm chuẩn hóa tên nhà máy để dễ so sánh
def normalize_plant_name(name):
    if not isinstance(name, str):
        return str(name)
    
    # Chuyển tất cả về lowercase
    name = name.lower()
    
    # Loại bỏ dấu, ký tự đặc biệt, dấu cách thừa
    name = re.sub(r'[^\w\s]', ' ', name)  # Thay thế ký tự đặc biệt bằng dấu cách
    name = re.sub(r'\s+', ' ', name)      # Thay thế nhiều dấu cách liên tiếp bằng một dấu cách
    name = name.strip()                   # Xóa dấu cách ở đầu và cuối
    
    # Thay thế dấu - bằng dấu cách
    name = name.replace('-', ' ')
    
    return name

# Tạo bản sao để không ảnh hưởng đến dữ liệu gốc
df_check_copy = df_check_SonLa.copy()
final_df_copy = final_df.copy()

# Thêm cột tên nhà máy chuẩn hóa
df_check_copy['normalized_name'] = df_check_copy['NMTD'].apply(normalize_plant_name)
final_df_copy['normalized_name'] = final_df_copy['Nhà máy'].apply(normalize_plant_name)

# Lấy danh sách các cột thời gian
time_columns = [col for col in df_check_copy.columns if ':' in col]
print(f"Số lượng cột thời gian: {len(time_columns)}")

# Tạo dictionary để ánh xạ tên nhà máy giữa hai DataFrame
plant_mapping = {}

# Với mỗi nhà máy trong df_check, tìm nhà máy tương ứng trong final_df
for plant in df_check_copy['normalized_name'].unique():
    # Tìm tên gần đúng trong final_df
    matches = get_close_matches(plant, final_df_copy['normalized_name'].unique(), n=1, cutoff=0.8)
    if matches:
        plant_mapping[plant] = matches[0]

print(f"Số lượng nhà máy đã ánh xạ: {len(plant_mapping)}")

# Tạo DataFrame để lưu kết quả chênh lệch
difference_results = []

# Tính sự chênh lệch cho từng cặp nhà máy
for df_check_plant, final_df_plant in plant_mapping.items():
    # Lấy dữ liệu của nhà máy từ df_check
    df_check_data = df_check_copy[df_check_copy['normalized_name'] == df_check_plant]
    
    # Lấy dữ liệu của nhà máy từ final_df
    final_df_data = final_df_copy[final_df_copy['normalized_name'] == final_df_plant]
    
    # Với mỗi ngày trong df_check_data, tìm ngày tương ứng trong final_df_data
    for idx, check_row in df_check_data.iterrows():
        check_date = check_row.get('NGAY')  # Giả sử cột ngày là 'NGAY'
        
        # Tìm dòng trong final_df_data có cùng ngày
        matching_final_rows = final_df_data[final_df_data['Ngày'] == check_date]
        
        if not matching_final_rows.empty:
            final_row = matching_final_rows.iloc[0]
            
            # Tính sự chênh lệch cho từng cột thời gian
            for time_col in time_columns:
                if time_col in final_row.index:
                    check_value = check_row.get(time_col, 0)
                    final_value = final_row.get(time_col, 0)
                    
                    # Tính chênh lệch tuyệt đối và phần trăm
                    abs_diff = abs(final_value - check_value)

                    # Lưu kết quả
                    difference_results.append({
                        'Nhà máy df_check': check_row['NMTD'],
                        'Nhà máy final_df': final_row['Nhà máy'],
                        'Ngày': check_date,
                        'Thời gian': time_col,
                        'Giá trị df_check': check_value,
                        'Giá trị final_df': final_value,
                        'Chênh lệch': abs_diff,
                    })

# Tạo DataFrame từ kết quả
difference_df = pd.DataFrame(difference_results)

# Hiển thị kết quả
print("\nKết quả chênh lệch:")
print(difference_df.head())

# Thống kê chênh lệch
print("\nThống kê chênh lệch:")
print(f"Tổng số điểm dữ liệu so sánh: {len(difference_df)}")

# Kiểm tra xem DataFrame có dữ liệu không và các cột có tồn tại không
if not difference_df.empty and 'Chênh lệch' in difference_df.columns:
    print(f"Chênh lệch trung bình: {difference_df['Chênh lệch'].mean()}")
else:
    print("DataFrame rỗng hoặc cột 'Chênh lệch' không tồn tại")

# Lưu kết quả (tùy chọn)
# difference_df.to_excel('chenh_lech_nha_may.xlsx', index=False)
```

```python
# Tạo DataFrame kết quả với hai cột đầu từ difference_df
result_df = difference_df[['Nhà máy df_check', 'Nhà máy final_df']].drop_duplicates().reset_index(drop=True)
result_df['order'] = result_df.index
# Lấy danh sách các ngày duy nhất
unique_days = difference_df['Ngày'].unique()

# Thêm cột P từ final_df
# Tạo dictionary ánh xạ từ tên nhà máy sang giá trị P
plant_to_p = {}
if 'P.Đặt' in final_df.columns:
    for _, row in final_df.iterrows():
        plant_name = row['Nhà máy']
        p_value = row['P.Đặt']
        plant_to_p[plant_name] = p_value

# Thêm cột P vào result_df
result_df['P.Đặt'] = result_df['Nhà máy final_df'].map(plant_to_p)

# Thêm cột Ngày với các giá trị unique
# Vì mỗi nhà máy có thể có nhiều ngày, cần tạo ra các hàng mới cho mỗi cặp (nhà máy, ngày)
new_rows = []
for _, row in result_df.iterrows():
    for day in unique_days:
        new_row = row.copy()
        new_row['Ngày'] = day
        new_rows.append(new_row)

# Tạo DataFrame mới từ các hàng đã tạo
result_df = pd.DataFrame(new_rows)

# Tính chênh lệch trung bình cho mỗi nhà máy và mỗi ngày
avg_differences = {}
for _, row in difference_df.iterrows():
    plant_check = row['Nhà máy df_check']
    plant_final = row['Nhà máy final_df']
    day = row['Ngày']
    
    key = (plant_check, plant_final, day)
    
    if key not in avg_differences:
        avg_differences[key] = []
    
    avg_differences[key].append(row['Chênh lệch'])

# Tính giá trị trung bình và thêm vào DataFrame kết quả
result_df['Chênh lệch trung bình'] = result_df.apply(
    lambda row: np.mean(avg_differences.get((row['Nhà máy df_check'], row['Nhà máy final_df'], row['Ngày']), [0])),
    axis=1
)

# Sử dụng cột order để sắp xếp kết quả cuối cùng
result_df = result_df.sort_values(['order', 'Ngày'])
result_df = result_df.drop(columns=['order'])  # Loại bỏ cột order sau khi sắp xếp

# Hiển thị kết quả
print("DataFrame kết quả:")
print(result_df.head())
print(f"Số lượng dòng: {len(result_df)}")
```
