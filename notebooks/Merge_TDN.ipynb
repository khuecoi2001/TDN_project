```python
import dask.dataframe as dd
import pandas as pd

# file path file sản lượng năm
file_path_sanluong = r"C:\Khue\TDN\data\raw\sanluongipp-2021.xlsm"
xlsm = pd.ExcelFile(file_path)
sheet_names = xlsm.sheet_names

# Xử lý mỗi sheet và kết hợp
dfs = []
for sheet in sheet_names:
    # Đọc sheet thành pandas DataFrame
    df = pd.read_excel(file_path, sheet_name=sheet)
    # Chuyển đổi thành Dask DataFrame
    ddf = dd.from_pandas(df, npartitions=10)  # Chia thành 10 phần
    dfs.append(ddf)

# Ghép tất cả DataFrame
combined_df = dd.concat(dfs)
```

```python
# Chuyển đổi kết quả từ dask DataFrame sang pandas DataFrame
result_df = combined_df.compute()
# reset index
result_df.reset_index(drop=True, inplace=True)
result_df
```

```python
# chỉ giữ lại các cột MADIEMDO, STARTTIME, CS của result_df
result_df = result_df[["MADIEMDO", "STARTTIME", "CS"]]
result_df
```

```python
# Tôi muốn chia cột STARTTIME thành 2 cột là DATE và TIME
result_df["NGAY"] = result_df["STARTTIME"].dt.date
result_df["TIME"] = result_df["STARTTIME"].dt.time
result_df = result_df.drop(columns=["STARTTIME"])
result_df
```

```python
# Tạo một DataFrame mới từ new_data, với cột đầu tiên là các giá trị unique của cột 'MADIEMDO', cột thứ hai là 'DATE', các cột tiếp theo là giá trị của cột 'CS' tương ứng với 'TIME'
pivot_df = pd.pivot_table(
    result_df, values="CS", index=["MADIEMDO", "NGAY"], columns="TIME"
).reset_index()
pivot_df
```
