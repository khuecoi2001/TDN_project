# VISUALIZATION

```python
import folium
import pandas as pd
```

```python
DS_TDN = pd.read_excel(r"D:\Khuê\Khue\TDN\data\raw\DS_TTT_TDN.xlsx")
```

```python
DS_TDN_ver1 = DS_TDN[["DIEN_LUC", "TEN_NM", "KINHDO", "VIDO", "TD_THAMCHIEU"]].dropna()
```

```python
DS_TDN_ver2 = DS_TDN_ver1[DS_TDN_ver1["TD_THAMCHIEU"] == "HOA_BINH"]
```

```python
DS_TDN_ver2 = DS_TDN_ver2.reset_index(drop=True)
DS_TDN_ver2
```

```python
DS_TDN_plot = DS_TDN_ver2.values.tolist()
```

```python
DS_TD_THAMCHIEU = pd.read_excel(r"D:\Khuê\Khue\TDN\data\raw\DS_TD_Thamchieu.xlsx")
```

```python
DS_TD_THAMCHIEU = DS_TD_THAMCHIEU[DS_TD_THAMCHIEU["TEN_NM"] == "HOA_BINH"]
```

```python
DS_TD_THAMCHIEU_plot = DS_TD_THAMCHIEU.values.tolist()
```

```python
map = folium.Map(location=[19.1969325, 104.0331585], zoom_start=9, tiles='Esri.WorldImagery')

for point in DS_TDN_plot:
    icon_TDN = folium.features.CustomIcon(
        r"D:\Khuê\Khue\TDN\references\images\TDN.PNG", icon_size=(30, 30)
    )
    folium.Marker(
        [point[3], point[2]], icon=icon_TDN, popup=folium.Popup(point[1], show=True)
    ).add_to(map)

for point in DS_TD_THAMCHIEU_plot:
    icon_TDL = folium.features.CustomIcon(
        r"D:\Khuê\Khue\TDN\references\images\TD_THAMCHIEU.png", icon_size=(40, 40)
    )
    folium.Marker(
        [point[3], point[2]], icon=icon_TDL, popup=folium.Popup(point[0], show=True)
    ).add_to(map)

    folium.CircleMarker(
        [point[3], point[2]],
        radius=40,
        color="tomato",
        stroke=False,
        fill=True,
        fill_opacity=0.6,
        opacity=1,
    ).add_to(map)

map
```
